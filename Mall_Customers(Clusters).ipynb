Choosing the Annual Income Column and spending Score columns

```python
x=data.iloc[:,[3,4]].values
```

```python
print(x)
```

Choosing the number of clusters

WCSS->Within Clusters Sum of Squares

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)

  wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Optimum Number of Clusters =5

Training the kmeans clustering model

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
y=kmeans.fit_predict(x)
```

```python
print(y)
```

Visualize all the characters

```python
plt.figure(figsize=(8,8))
plt.scatter(x[y==0,0],x[y==0,1],s=50,c='green',label='cluster 1')
plt.scatter(x[y==1,0],x[y==1,1],s=50,c='red',label='cluster 2')
plt.scatter(x[y==2,0],x[y==2,1],s=50,c='purple',label='cluster 3')
plt.scatter(x[y==3,0],x[y==3,1],s=50,c='black',label='cluster 4')
plt.scatter(x[y==4,0],x[y==4,1],s=50,c='blue',label='cluster 5')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='cyan',label='centroids')
plt.title('Customer Group')
plt.xlabel('Annual Income')
plt.show()
```
